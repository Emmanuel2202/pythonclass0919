```python
import numpy as np
import pandas as pd
import requests
```

```python
url_link = 'https://pokemondb.net/pokedex/all'
r = requests.get(url_link,headers ={'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'})
read_html_pandas_data = pd.read_html(r.text)[0]
df= pd.DataFrame(read_html_pandas_data)
df.head()
```

```python
df.shape
```

```python
df.to_csv('pokeman.csv',index=False)
```

```python
df.to_excel('pokeman.xlsx',sheet_name='pokedex',index=False)
```

```python
df['Type'].unique()
```

```python
with pd.ExcelWriter('pokemantype.xlsx',engine ='xlsxwriter') as writer:
    df[df['Type']=='Dark Ice'].to_excel(writer,sheet_name='DarkIce',index=False,engine= 'xlsxwriter')
    df[df['Type']=='Bug Water'].to_excel(writer,sheet_name='BugWater',index=False,engine= 'xlsxwriter')
    df[df['Type']=='Steel Ground'].to_excel(writer,sheet_name='SteelGround',index=False,engine= 'xlsxwriter')
    df[df['Type']=='Fire Fighting'].to_excel(writer,sheet_name='FireFighting',index=False,engine= 'xlsxwriter')
```
